```python
%%sql
#Total Quantity Sold Per Product
select p.product_name,p.product_id,p.category,sum(o.quantity) as total_sold
from products p
 join order_items o 
on p.product_id=o.product_id
group by p.product_id,p.category
;
```

```python
%%sql
# which product spld the most
select p.product_id,p.product_name,p.category,sum(o.quantity) as Top_product
from products p
join order_items o 
on p.product_id=o.product_id
group by p.product_id,p.product_name
order by sum(o.quantity) desc
limit 1 
```

```python
%%sql
# Top 5 Products
select p.product_id,p.product_name,sum(o.quantity) as top_5_product
from products p 
join order_items o 
on p.product_id=o.product_id 
group by p.product_id,p.product_name
order by top_5_product desc 
limit 5 ;
```

```python
%%sql
# Category with highest number of products sold
select p.category,sum(o.quantity) as total_quantity 
from products p 
join order_items o on p.product_id=o.product_id 
group by p.category 
order by total_quantity desc
limit 1;
```

```python
%%sql
#Total revenue generated
select sum(p.price*o.quantity) as total_revenue 
from products p 
join order_items o 
on p.product_id=o.order_id 
```

```python
%%sql
# Total revenue generated per category 
select p.category,sum(p.price*o.quantity) as total_quantity
from products p 
join order_items o 
on o.product_id=p.product_id 
group by p.category 
order by total_quantity desc;
```

```python
%%sql
# Revenue per year
select year(o.order_date)as year,sum(p.price*oi.quantity) as price
from products p 
join order_items  oi on p.product_id=oi.product_id 
join orders o on o.order_id=oi.order_id
group by year(o.order_date)
order by price desc;
```

```python
%%sql
#Revenue per month
select year(o.order_date) as year,month(o.order_date) as month,sum(p.price*oi.quantity) as price 
from products p 
join order_items oi on p.product_id=oi.product_id 
join orders o on o.order_id=oi.order_id
group by year(o.order_date),month(o.order_date) 
order by year,month desc;
```

```python
%%sql
# which customer placed the most orders
select c.customer_id,c.customer_firstname,count(o.order_date) as total_orders
from customers c 
join orders o on c.customer_id=o.customer_id 
group by c.customer_id,c.customer_firstname 
order by total_orders desc
limit 10;
```

```python
%%sql
#Which Customer Bought the Highest Total Quantity?
select c.customer_id,concat(c.customer_firstname,' ',c.customer_lastname) as full_name,sum(oi.quantity) as high
from customers c 
join orders o on c.customer_id =o.customer_id
join order_items oi  on o.order_id =oi.order_id 
group by c.customer_id,full_name
order by high desc 
limit 1;
```

```python
%%sql
#Which Customer Generated the Highest Revenue?
select c.customer_id,concat(c.customer_firstname,' ',c.customer_lastname)as Name,sum(p.price*oi.quantity) as high_price
from customers c 
join orders o on c.customer_id=o.customer_id 
join order_items oi on o.order_id=oi.order_id 
join products p on p.product_id =oi.product_id 
group by c.customer_id,name
order by high_price desc 
limit 1;
```

```python
%%sql
#Average order value
with total_revenue as (
    select o.order_id,sum(p.price*oi.quantity) as revenue 
    from orders o join order_items oi on o.order_id=oi.order_id
    join products p on  p.product_id=oi.product_id
    group by o.order_id
)
select round(avg(revenue),2) from total_revenue;
```

```python
%%sql
#Orders That Contain More Than 3 Products
select o.order_id,
sum(oi.quantity) as total_items
from orders o 
join order_items oi on o.order_id=oi.order_id 
group by o.order_id 
having total_items>3;
```
