```python
%%sql
CREATE DATABASE ocommerce_db;
USE ocommerce_db;
```

```python
%%sql
CREATE TABLE customers (
    customer_id INT PRIMARY KEY,
    name VARCHAR(50),
    email VARCHAR(100),
    city VARCHAR(50),
    signup_date DATE
);
```

```python
%%sql
CREATE TABLE products (
    product_id INT PRIMARY KEY,
    product_name VARCHAR(100),
    category VARCHAR(50),
    price DECIMAL(10,2)
);
```

```python
%%sql
CREATE TABLE orders (
    order_id INT PRIMARY KEY,
    customer_id INT,
    order_date DATE,
    total_amount DECIMAL(10,2),
    FOREIGN KEY (customer_id) REFERENCES customers(customer_id)
);
```

```python
%%sql
CREATE TABLE order_items (
    order_item_id INT PRIMARY KEY,
    order_id INT,
    product_id INT,
    quantity INT,
    price DECIMAL(10,2),
    FOREIGN KEY (order_id) REFERENCES orders(order_id),
    FOREIGN KEY (product_id) REFERENCES products(product_id)
);
```

```python
%%sql
#Find total money spent by each customer.
select c.customer_id,sum(oi.quantity*oi.price) as total_spent
from customers c
join orders o on c.customer_id=o.customer_id 
join order_items oi on o.order_id=oi.order_id
group by c.customer_id
order by total_spent;
```

```python
%%sql
##Find top 3 customers who spent the most money
select c.customer_id,sum(oi.quantity*oi.price) as total_spent
from customers c
join orders o on c.customer_id=o.customer_id 
join order_items oi on o.order_id=oi.order_id
group by c.customer_id
order by total_spent
limit 3;
```

```python
%%sql
# Find most sold product.
select product_id,sum(quantity) 
from order_items
group by product_id 
order by sum(quantity) desc;
```

```python
%%sql
#Find customers who never placed any order.
select c.customer_id 
from customers c 
join orders o 
on c.customer_id=o.order_id 
where o.order_id is null;
```

```python
%%sql
#Find revenue generated by each product category.
select p.category,sum(oi.price*oi.quantity) as total
from products p 
join order_items oi on p.product_id=oi.product_id 
group by p.category
```

```python
%%sql
# Top Product in Each Category
select p.category,sum(oi.price*oi.quantity) as total 
from products p 
join order_items oi on p.product_id=oi.product_id 
group by p.category 
order by total desc;
```

```python

%%sql
SELECT category, product_name, revenue
FROM (
    SELECT p.category,
           p.product_name,
           SUM(oi.quantity * oi.price) AS revenue,
           RANK() OVER (PARTITION BY p.category ORDER BY SUM(oi.quantity * oi.price) DESC) AS rnk
    FROM order_items oi
    JOIN products p
    ON oi.product_id = p.product_id
    GROUP BY p.category, p.product_name
) t
WHERE rnk = 1;
```

```python
%%sql
# Customers With More Than 1 Order
select c.customer_id,c.name
from customers c 
join orders o on c.customer_id=o.customer_id 
group by c.customer_id
having count(o.order_id)>1
```
