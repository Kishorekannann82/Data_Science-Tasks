```python
%%sql
# total revenue generated
select round(sum(fare),2) as Total_revenue from trips 
where status="Completed";
```

```python
%%sql
# total number of trips per city
select count(t.trip_id) as total_number_trips
from Trips t 
join Riders r on t.rider_id=r.rider_id;
```

```python
%%sql
# Average trip distance
select round(avg(distance_km),2) as Average_Trip_Distance
from Trips;
```

```python
%%sql
# top 5 drivers with highest average driver_rating
select d.driver_id,d.driver_name ,avg(r.driver_rating) as avg_rating
from drivers d 
join trips t on d.driver_id=t.driver_id 
join ratings r on t.trip_id=r.trip_id
group by d.driver_id,d.driver_name
order by avg_rating
limit 5;
```

```python
%%sql
#Total Cancellation Percentage
with trip_counts as(
    select count(*) as total_trip,
    sum(case when status= "Cancelled" then 1 else 0 end ) as cancelled_trip
    from trips
    )
select total_trip,cancelled_trip,round(
(cancelled_trip*100/total_trip),2) as Cancellation_Percentage
from trip_counts
```

```python
%%sql 
with trip_counts as (
    select count(*) as total_trips,
    sum(case when status='Completed' then 1 else 0 end) as Completed_Trips
    from trips
    )
select total_trips,completed_trips,
round((completed_trips*100/total_trips),2) as Completed_Precentage
from trip_Counts;
```

```python
%%sql
#Most Prefered Payments
select payment_method,count(payment_method) as Majority_Payment
from Payments
group by payment_method;
```

```python
%%sql
#top 3 highest earning drivers
select d.driver_id,d.driver_name,sum(t.fare) as total_fare
from drivers d 
join trips t on  d.driver_id=t.driver_id
group by d.driver_id,d.driver_name 
order by total_fare desc
limit 3 ;
```

```python
%%sql
#Riders with more than 3 trips
select r.rider_id,r.rider_name,count(t.trip_id) as total
from riders r 
join trips t on r.rider_id=t.rider_id 
group by r.rider_id,r.rider_name
having count(t.trip_id)>=3;
```

```python
%%sql
#Drivers Who Have Never Received Rating Below 4 (Rating>4 only)
select d.driver_id,d.driver_name
from drivers d 
join trips t on t.driver_id=d.driver_id 
join ratings r on t.trip_id=r.trip_id
group by d.driver_id,d.driver_name
having min(r.driver_rating)>=4;
```

```python
%%sql
#City With Highest Revenue->Rider
select r.city,sum(t.fare) as revenue
from Riders r 
join trips t on r.rider_id=t.rider_id 
group by r.city 
order by revenue desc 
limit 1;
```

```python
%%sql 
#City with Lowest Revenue
select r.city,sum(t.fare) as revenue 
from riders r 
join trips t on t.rider_id=r.rider_id 
group by r.city 
order by revenue 
limit 1;
```

```python
%%sql
# Drivers who received less rating
select d.driver_id,d.driver_name
from drivers d 
join trips t on d.driver_id=t.driver_id 
join ratings r on t.trip_id=r.trip_id 
group by d.driver_id,d.driver_name
having min(r.rider_rating)<=3;
```

```python
%%sql 
#Lowest Earning Drivers
select d.driver_id,d.driver_name,sum(t.fare) as fare
from drivers d 
join trips t on t.driver_id=d.driver_id
group by d.driver_id,d.driver_name
order by fare asc
limit 3;
```
