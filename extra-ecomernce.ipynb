```python
%%sql
use ecommerce;
```

```python
%%sql
#Total revenue for each customer
select c.customer_id,c.customer_firstname,sum(p.price*oi.quantity) as revenue 
from customers c 
join orders o on c.customer_id=o.customer_id
join order_items oi on  o.order_id=oi.order_id 
join products p on oi.product_id=p.product_id
group by c.customer_id,c.customer_firstname
order by revenue desc;
```

```python
%%sql
select year(order_date) as year,count(year(order_date)) as year_count
from orders
group by year(order_date);
```
