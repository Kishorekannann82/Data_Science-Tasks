```python
"""
Air Passengers Time Series Forecasting
=======================================
Train : 1949-01 to 1954-12  (72 months)
Test  : 1955-01 to 1960-12  (72 months)
Models: ARIMA, SARIMA, Prophet

Install dependencies:
    pip install pandas numpy matplotlib statsmodels prophet scikit-learn
"""

import warnings
warnings.filterwarnings("ignore")

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.statespace.sarimax import SARIMAX
from statsmodels.tsa.stattools import adfuller
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from sklearn.metrics import mean_absolute_error, mean_squared_error
from prophet import Prophet

# ─────────────────────────────────────────────
# 1. LOAD DATA
# ─────────────────────────────────────────────
# Using the classic AirPassengers dataset (built into seaborn / available via URL)
url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/airline-passengers.csv"
try:
    df = pd.read_csv(url, parse_dates=["Month"], index_col="Month")
    df.columns = ["Passengers"]
    print("✓ Loaded dataset from URL")
except Exception:
    # Fallback: inline data
    raw = [
        112,118,132,129,121,135,148,148,136,119,104,118,
        115,126,141,135,125,149,170,170,158,133,114,140,
        145,150,178,163,172,178,199,199,184,162,146,166,
        171,180,193,181,183,218,230,242,209,191,172,194,
        196,196,236,235,229,243,264,272,237,211,180,201,
        204,188,235,227,234,264,302,293,259,229,203,229,
        242,233,267,269,270,315,364,347,312,274,237,278,
        284,277,317,313,318,374,413,405,355,306,271,306,
        315,301,356,348,355,422,465,467,404,347,305,336,
        340,318,362,348,363,435,491,505,404,359,310,337,
        360,342,406,396,420,472,548,559,463,407,362,405,
        417,391,419,461,472,535,622,606,508,461,390,432,
    ]
    idx = pd.date_range("1949-01", periods=144, freq="MS")
    df = pd.DataFrame({"Passengers": raw}, index=idx)
    print("✓ Loaded inline dataset")

df.index.freq = "MS"
print(df.head())
print(f"\nTotal observations: {len(df)}")

# ─────────────────────────────────────────────
# 2. TRAIN / TEST SPLIT
# ─────────────────────────────────────────────
train = df["1949":"1954"]   # 72 months
test  = df["1955":"1960"]   # 72 months

print(f"\nTrain: {train.index[0].date()} → {train.index[-1].date()} ({len(train)} obs)")
print(f"Test : {test.index[0].date()} → {test.index[-1].date()}  ({len(test)} obs)")

# ─────────────────────────────────────────────
# 3. EXPLORATORY – ADF Stationarity Test
# ─────────────────────────────────────────────
print("\n── ADF Stationarity Test (original series) ──")
adf_result = adfuller(train["Passengers"])
print(f"  ADF Statistic : {adf_result[0]:.4f}")
print(f"  p-value       : {adf_result[1]:.4f}")
print(f"  Stationary?   : {'Yes' if adf_result[1] < 0.05 else 'No – differencing needed'}")

# ─────────────────────────────────────────────
# 4. HELPER: METRICS
# ─────────────────────────────────────────────
def metrics(actual, predicted, model_name):
    mae  = mean_absolute_error(actual, predicted)
    rmse = np.sqrt(mean_squared_error(actual, predicted))
    mape = np.mean(np.abs((actual - predicted) / actual)) * 100
    print(f"\n── {model_name} Metrics ──")
    print(f"  MAE  : {mae:.2f}")
    print(f"  RMSE : {rmse:.2f}")
    print(f"  MAPE : {mape:.2f}%")
    return {"Model": model_name, "MAE": round(mae, 2), "RMSE": round(rmse, 2), "MAPE (%)": round(mape, 2)}

results = []

# ─────────────────────────────────────────────
# 5. MODEL 1 – ARIMA(2,1,0)
# ─────────────────────────────────────────────
print("\n══════════════════════════════")
print("  MODEL 1: ARIMA(2,1,0)")
print("══════════════════════════════")

arima_model  = ARIMA(train["Passengers"], order=(2, 1, 0))
arima_fit    = arima_model.fit()
print(arima_fit.summary())

# Forecast 72 steps ahead
arima_fc     = arima_fit.forecast(steps=len(test))
arima_fc.index = test.index

results.append(metrics(test["Passengers"], arima_fc, "ARIMA(2,1,0)"))

# ─────────────────────────────────────────────
# 6. MODEL 2 – SARIMA(2,1,0)(1,1,1,12)
# ─────────────────────────────────────────────
print("\n══════════════════════════════")
print("  MODEL 2: SARIMA(2,1,0)(1,1,1)12")
print("══════════════════════════════")

sarima_model = SARIMAX(
    train["Passengers"],
    order=(2, 1, 0),
    seasonal_order=(1, 1, 1, 12),
    enforce_stationarity=False,
    enforce_invertibility=False,
)
sarima_fit   = sarima_model.fit(disp=False)
print(sarima_fit.summary())

sarima_fc    = sarima_fit.forecast(steps=len(test))
sarima_fc.index = test.index

results.append(metrics(test["Passengers"], sarima_fc, "SARIMA(2,1,0)(1,1,1)12"))

# ─────────────────────────────────────────────
# 7. MODEL 3 – Prophet
# ─────────────────────────────────────────────
print("\n══════════════════════════════")
print("  MODEL 3: Prophet")
print("══════════════════════════════")

# Prophet requires columns: ds (datetime), y (value)
prophet_train = train.reset_index().rename(columns={"Month": "ds", "Passengers": "y"})
# If index column is unnamed after reset:
if "ds" not in prophet_train.columns:
    prophet_train = train.reset_index()
    prophet_train.columns = ["ds", "y"]

prophet_model = Prophet(
    yearly_seasonality=True,
    weekly_seasonality=False,
    daily_seasonality=False,
    seasonality_mode="multiplicative",  # works better for growing seasonal amplitude
    changepoint_prior_scale=0.05,
)
prophet_model.fit(prophet_train)

future     = prophet_model.make_future_dataframe(periods=len(test), freq="MS")
forecast   = prophet_model.predict(future)
prophet_fc = forecast.set_index("ds")["yhat"].loc[test.index]

results.append(metrics(test["Passengers"], prophet_fc, "Prophet"))

# ─────────────────────────────────────────────
# 8. SUMMARY TABLE
# ─────────────────────────────────────────────
print("\n══════════════════════════════════════")
print("  MODEL COMPARISON SUMMARY")
print("══════════════════════════════════════")
summary_df = pd.DataFrame(results).set_index("Model")
print(summary_df.to_string())
best = summary_df["MAPE (%)"].idxmin()
print(f"\n🏆 Best model by MAPE: {best}")

# ─────────────────────────────────────────────
# 9. VISUALISATION
# ─────────────────────────────────────────────
fig, axes = plt.subplots(3, 1, figsize=(14, 16), sharex=False)
fig.suptitle("Air Passenger Forecasting — 1949 to 1960", fontsize=16, fontweight="bold", y=0.98)

palette = {
    "Actual (Train)": "#94a3b8",
    "Actual (Test)":  "#f8fafc",
    "ARIMA":          "#f97316",
    "SARIMA":         "#22d3ee",
    "Prophet":        "#a78bfa",
}

forecast_sets = [
    ("ARIMA(2,1,0)",            arima_fc,   "ARIMA"),
    ("SARIMA(2,1,0)(1,1,1)12",  sarima_fc,  "SARIMA"),
    ("Prophet",                 prophet_fc, "Prophet"),
]

for ax, (title, fc, key) in zip(axes, forecast_sets):
    ax.plot(train.index, train["Passengers"], color=palette["Actual (Train)"],
            linewidth=1.8, label="Train (actual)", alpha=0.8)
    ax.plot(test.index,  test["Passengers"],  color=palette["Actual (Test)"],
            linewidth=2,   label="Test (actual)")
    ax.plot(test.index,  fc,                  color=palette[key],
            linewidth=2.2, linestyle="--", label=f"{key} Forecast")
    ax.axvline(pd.Timestamp("1955-01-01"), color="#475569", linestyle=":", linewidth=1.2)
    ax.text(pd.Timestamp("1955-02-01"), ax.get_ylim()[0] * 1.02, "Forecast →",
            fontsize=9, color="#64748b")
    ax.set_title(title, fontsize=13, fontweight="bold", pad=8)
    ax.set_ylabel("Passengers (thousands)", fontsize=10)
    ax.legend(fontsize=10)
    ax.xaxis.set_major_locator(mdates.YearLocator())
    ax.xaxis.set_major_formatter(mdates.DateFormatter("%Y"))
    ax.grid(True, alpha=0.2, linestyle="--")
    # Annotate MAPE
    m = summary_df.loc[[k for k in summary_df.index if key in k][0], "MAPE (%)"]
    ax.annotate(f"MAPE: {m}%", xy=(0.97, 0.92), xycoords="axes fraction",
                fontsize=11, ha="right", color=palette[key],
                bbox=dict(boxstyle="round,pad=0.3", fc="#0f172a", ec=palette[key], alpha=0.8))

plt.tight_layout()
plt.savefig("air_passengers_forecast.png", dpi=150, bbox_inches="tight",
            facecolor="#0a0f1e")
plt.show()
print("\n✓ Plot saved as air_passengers_forecast.png")

# ─────────────────────────────────────────────
# 10. COMBINED PLOT (all models on one chart)
# ─────────────────────────────────────────────
fig2, ax = plt.subplots(figsize=(14, 6))
ax.set_facecolor("#0f172a")
fig2.patch.set_facecolor("#0a0f1e")

ax.plot(train.index, train["Passengers"], color="#94a3b8", linewidth=1.5, label="Train")
ax.plot(test.index,  test["Passengers"],  color="#f8fafc", linewidth=2,   label="Actual (Test)")
ax.plot(test.index,  arima_fc,            color="#f97316", linewidth=1.8, linestyle="--", label="ARIMA")
ax.plot(test.index,  sarima_fc,           color="#22d3ee", linewidth=1.8, linestyle="--", label="SARIMA")
ax.plot(test.index,  prophet_fc,          color="#a78bfa", linewidth=1.8, linestyle="--", label="Prophet")
ax.axvline(pd.Timestamp("1955-01-01"), color="#475569", linestyle=":", linewidth=1.5, label="Train/Test Split")

ax.set_title("Air Passengers — All Models Combined", fontsize=14, fontweight="bold", color="#f8fafc", pad=12)
ax.set_xlabel("Year", fontsize=11, color="#94a3b8")
ax.set_ylabel("Passengers (thousands)", fontsize=11, color="#94a3b8")
ax.tick_params(colors="#64748b")
ax.xaxis.set_major_locator(mdates.YearLocator())
ax.xaxis.set_major_formatter(mdates.DateFormatter("%Y"))
ax.grid(True, alpha=0.15, linestyle="--", color="#334155")
ax.legend(fontsize=10, facecolor="#1e293b", edgecolor="#334155", labelcolor="#e2e8f0")

plt.tight_layout()
plt.savefig("air_passengers_combined.png", dpi=150, bbox_inches="tight", facecolor="#0a0f1e")
plt.show()
print("✓ Combined plot saved as air_passengers_combined.png")

# ─────────────────────────────────────────────
# 11. FORECAST TABLE (first 12 months)
# ─────────────────────────────────────────────
print("\n── Forecast Table: Jan 1955 – Dec 1955 ──")
forecast_table = pd.DataFrame({
    "Actual"  : test["Passengers"][:12].values,
    "ARIMA"   : arima_fc[:12].round().astype(int).values,
    "SARIMA"  : sarima_fc[:12].round().astype(int).values,
    "Prophet" : prophet_fc[:12].round().astype(int).values,
}, index=test.index[:12].strftime("%Y-%m"))
forecast_table["ARIMA_err%"]   = ((forecast_table["ARIMA"]   - forecast_table["Actual"]) / forecast_table["Actual"] * 100).round(1)
forecast_table["SARIMA_err%"]  = ((forecast_table["SARIMA"]  - forecast_table["Actual"]) / forecast_table["Actual"] * 100).round(1)
forecast_table["Prophet_err%"] = ((forecast_table["Prophet"] - forecast_table["Actual"]) / forecast_table["Actual"] * 100).round(1)
print(forecast_table.to_string())
```
